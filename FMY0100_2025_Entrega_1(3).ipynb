## 11. Fase 2 - Existencia de atípicos


`````python

def buscar_atipicos(data : pd.DataFrame, columna : str) -> pd.DataFrame:
  """
  Busca valores atípicos en una columna.
  """
  # Calcular los límites
  Q1 = data[columna].quantile(0.25)
  Q3 = data[columna].quantile(0.75)
  # Calcula rango intercuartilico
  IQR = Q3 - Q1
  limite_inferior = Q1 - 1.5 * IQR
  limite_superior = Q3 + 1.5 * IQR

  # Filtrar outliers
  return data[(data[columna] < limite_inferior) | (data[columna] > limite_superior)]

def obtener_cantidad_atipicos(data : pd.DataFrame, columnas : np.array) -> dict:
  """
  Obtiene la cantidad de atípicos por cada columna.
  """
  total_atipicos = {}
  for columna in data[columnas]:
    atipicos = buscar_atipicos(data, columna)
    total_atipicos[columna] = atipicos.shape[0]
  return total_atipicos
atipicos_por_columna = obtener_cantidad_atipicos(df, df.describe().columns)
atipicos_por_columna



`````

## 16. Fase 2 - Aplicación CORRECTA de prueba ANOVA






Variable independiente discreta: Map

Variable dependiente discreta : MatchKills

`````python
# Muestra aleatoria de 100 registros del DataFrame
# Establece una semilla para extraer siempre la misma muestra
random.seed(30)
muestra_df = df.sample(n=100, random_state=30).reset_index(drop=True)

# Dos variables de interés: Map y MatchKills
muestra_df = muestra_df[['Map', 'MatchKills']]

# Crear tabla con mapas y estadísticas de MatchKills
groups = muestra_df.groupby('Map').agg({
    'MatchKills': ['count', 'mean', 'std', 'min', 'max']
}).round(2)

# Aplanar los nombres de las columnas
groups.columns = ['Count', 'Media_MatchKills', 'DesvEst_MatchKills', 'Min_MatchKills', 'Max_MatchKills']
groups = groups.reset_index()

# Ordenar por Count de mayor a menor
groups_ordenado = groups.sort_values(by='Count', ascending=False).reset_index(drop=True)

# Visualizar la tabla ordenada
groups_ordenado
`````

### Definición de hipótesis
**H₀** = Todos los mapas tienen el mismo promedio de MatchKills

**H₁** = Al menos un mapa tiene diferente promedio de MatchKills

**α** = 0.05 (Grados de libertad ya calculados)


`````python
# Crea la tabla principal de ANOVA
data = [['Between Groups', '', '', '', '', '', ''],
        ['Within Groups', '', '', '', '', '', ''],
        ['Total', '', '', '', '', '', '']]

anova_table = pd.DataFrame(data, columns=['Source of Variation', 'SS', 'df', 'MS', 'F', 'P-value', 'F crit'])
anova_table.set_index('Source of Variation', inplace=True)

# Calcula ENTRE los grupos
x_bar = muestra_df['MatchKills'].mean()
SSTR = muestra_df.groupby('Map').count() * (muestra_df.groupby('Map').mean() - x_bar)**2
anova_table.loc['Between Groups', 'SS'] = SSTR['MatchKills'].sum()

# Calcula EN los grupos
SSE = (muestra_df.groupby('Map').count() - 1) * muestra_df.groupby('Map').std()**2
anova_table.loc['Within Groups', 'SS'] = SSE['MatchKills'].sum()

# Calcula el total
SSTR_total = SSTR['MatchKills'].sum() + SSE['MatchKills'].sum()
anova_table.loc['Total', 'SS'] = SSTR_total

# Actualiza los grados de libertad
anova_table.loc['Between Groups', 'df'] = muestra_df['Map'].nunique() - 1
anova_table.loc['Within Groups', 'df'] = muestra_df.shape[0] - muestra_df['Map'].nunique()
anova_table.loc['Total', 'df'] = muestra_df.shape[0] - 1

# Calcula la media de los cuadrados
anova_table['MS'] = anova_table['SS'] / anova_table['df']

# Calcula F
F = anova_table.loc['Between Groups', 'MS'] / anova_table.loc['Within Groups', 'MS']
anova_table.loc['Between Groups', 'F'] = F

# p-value
anova_table.loc['Between Groups', 'P-value'] = 1 - stats.f.cdf(F, anova_table.loc['Between Groups', 'df'], anova_table.loc['Within Groups', 'df'])

# F critico
alpha = 0.05
anova_table.loc['Between Groups', 'F crit'] = stats.f.ppf(1-alpha, anova_table.loc['Between Groups', 'df'], anova_table.loc['Within Groups', 'df'])

# Mostrar tabla ANOVA
anova_table
`````

`````python
# Resultado final
print("\n","-"*80)
print("El enfoque del valor crítico para la prueba de hipótesis en la regla de decisión")
print("Se rechaza H₀, cuando F > Fα (F crit)")
conclusion = "No se pudo rechazar la hipótesis nula (H₀)."
if anova_table['F']['Between Groups'] > anova_table['F crit']['Between Groups']:
    conclusion = "Se rechaza la hipótesis nula (H₀)."
print("F-score es:", anova_table['F']['Between Groups'], " y valor critico es:", anova_table['F crit']['Between Groups'])
print(conclusion)
`````

### Conclusión

No se pudo rechazar la hipótesis nula (H₀).

Los resultados sugieren que el mapa en el que se juega no influye significativamente en el promedio de kills (al menos con la muestra y nivel de significancia considerado)
